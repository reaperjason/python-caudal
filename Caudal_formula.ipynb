# cálculo de caudal con la fórmula
## Caudal ecológico (Qe) = (0.8 × Caudal medio) + (0.2 × Caudal máximo)

```python
%matplotlib notebook
import pandas as pd
datos_caudales = pd.read_excel('datos.xlsx', index_col=0, parse_dates=True)
```

```python
datos_caudales['Q   ( m3/s)'].plot()
```

```python
solo_caudales=datos_caudales['Q   ( m3/s)']
caudal_medio= sum(solo_caudales)/len(solo_caudales)
print(caudal_medio)
```

```python
caudal_maximo = max(solo_caudales)
print(caudal_maximo)
```

```python
caudal_ecologico = (0.8 * caudal_medio)+(0.2 * caudal_maximo)
print(caudal_ecologico)
```
